```python
import pandas as pd
df=pd.read_csv('/content/Mall_Customers (1).csv')
import warnings
warnings.filterwarnings('ignore')
df
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Genre']=le.fit_transform(df['Genre'])
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import normalize
data=normalize(df)
data
```

```python
newdata=pd.DataFrame(data,columns=df.columns)
newdata
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.title('DENDROGRAM')
dend=shc.dendrogram(shc.linkage(newdata,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
cluster.fit_predict(newdata)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(newdata['Annual Income (k$)'],newdata['Spending Score (1-100)'],c=cluster.labels_)
```
